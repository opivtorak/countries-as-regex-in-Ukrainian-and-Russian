```python
import pandas as pd
```

## Технічна колонка з назвами країн малими буквами

```python
df=pd.read_excel("countries_to_prepare.xlsx")
df["Link_lower"]=df["Link"].str.lower()
```

## Масив з назвою країни представленою останнім словом

```python
array=[]

for cell in df["Link_lower"]:
    dd=cell.rsplit(' ', 1)[-1]
    array.append(dd)
    
df['Last_word']=array
df
```

## Результат

```python
df["first_letters"]=df["Last_word"].str[:-2]

df["processed"]=df["first_letters"]+'([А-Яа-яЁёЇїЄєЭэЪъ]+)?'
final=df[["Link", "processed"]]
final.to_excel('countries_to_finish_manually.xlsx')
               
final
```
